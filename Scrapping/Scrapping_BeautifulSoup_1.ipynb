# BeautifulSoup 모듈을 이용한 스크래핑 실습


### Beautifulsoup 모듈
- 파싱을 도와주는 python 라이브러리
- 파싱(Parsing)이란 가공되지 않은 데이터에서 원하는 특정한 문자열을 빼내는 작업

```python
# 관련 모듈 임포트
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
# url 접근후 html 데이터 획득
# 샘플데이터url : http://pythonscraping.com/pages/warandpeace.html
url = 'http://pythonscraping.com/pages/warandpeace.html'

# 객체 색성
html = urlopen(url)

# html 소스 저장
html_str = html.read()

print(html_str)
```

```python
# soup 객체 생성 : 파싱
soup = bs(html_str, 'html.parser')
print(soup)
```

### 도트(.)연산자를 이용한 태그 찾기
- soup객체.태그1.태그2...
- next_sibling : 태그의 다음 요소 정보 추출
- string 속성 : 태그의 텍스트만 추출

```python
# soup 객체안의 h1 태그 요소 찾기
print(soup.h1)
print(soup.h1.string)
print(soup.h1.next_sibling)
print(soup.h1.next_sibling.next_sibling)
print(soup.h1.next_sibling.next_sibling.string)
```

```python
# soup 객체안의 div > span 찾기 
# print(soup.div)
print(soup.div.span)
print('-'*50)
print(soup.div.span.string)
```

### 아이디로 찾기 : find()
- soup객체.find(id='아이디명') => 해당아이디의 태그 
- string => 텍스트

```python
# id값이 text 
print(soup.find(id='text'))
print('\n')
print(soup.find(id='text').string) # None
print('\n')
print(soup.find(id='text').span.string)
```

### 클래스로 찾기 : find_all()

- soup객체.find_all(class_='클래스명') => 해당 클래스의 태그들. 리스트
- string => 텍스트 
- soup객체.find_all(class_='클래스명')[인덱스].string

```python
# class 값이 red 인 태그 찾기 
red_list = soup.find_all(class_='red')
print('갯수 => ', len(red_list))
print(red_list)
```

```python
# red 클래스가 적용된 마지막 태그의 텍스트만 출력 
# print(red_list.string) # AttributeError   
print(red_list[-1].string)
```

### 같은 태그명 찾기 : find_all()

- soup객체.find_all('태그명') =>  리스트
- string => 텍스트 
- soup객체.find_all('태그명')[인덱스].string

```python
# span 태그들 찾기
span_tag_list = soup.find_all('span')
print('갯수 => ', len(span_tag_list))
print('\n')
print(span_tag_list[0])
```

### 글자로 찾기 : find_all()

- soup객체.find_all(text='문자열') => 리스트
- 문자열은 2개이상의 어구

```python
name_list = soup.find_all(text='the prince')
print('갯수 => ', len(name_list))
print(name_list)
```

### 속성 값으로 찾기 : find_all()

- soup객체.find_all(attrs={속성:값}) => 리스트

```python
# class 값이 green인 태그 찾기
green_list = soup.find_all(attrs={'class':'green'})
print('갯수 => ', len(green_list))
print(green_list[-1])
```

### 선택자로 찾기 : select(), select_one()

- soup객체.select(선택자) => 리스트
- soup객체.select_one(선택자) => 태그

```python
# 선택자 정의
# target_selector = '#text > p:nth_child(3)' # 인식 못함
target_selector = '#text > p:nth-of-type(3)'
print(soup.select(target_selector))

# red 클래스가 적용된 span 선택자
target_selector = 'span.red'
print(soup.select(target_selector))
```
