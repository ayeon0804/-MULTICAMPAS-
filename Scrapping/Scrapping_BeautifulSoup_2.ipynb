# 네이버 뉴스토픽 인기 순위

- url : https://search.naver.com/search.naver?where=nexearch
- 뉴스 / 연예스포츠 1~10위
- 제목과 url

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch'
html = urlopen(url)
html_str = html.read()

soup = bs(html_str, 'html.parser')
print(soup)
```

```python
target_selector = 'ol.keyword_list > li.bx'
print(len(target_selector))
```

```python
li_list = soup.select(target_selector)
print(li_list)
print('\n')
print(li_list[0])
```

```python
# 속성값 추출은? attrs['속성']
base_url = 'https://search.naver.com/search.naver'
print('순위 => ',li_list[0].em.string)
print('하이퍼링크 => ', base_url + li_list[0].a.attrs['href'])
print('제목 => ',li_list[0].a.span.string)
```

```python
# for문을 이용해서 리스트로 저장
base_url = 'https://search.naver.com/search.naver'
naver_keyword_list = []
for li in li_list:
    score = li.em.string
    k_url = base_url + li.a.attrs['href']
    title = li.a.span.string
    naver_keyword_list.append([score, title, k_url])

for row in naver_keyword_list:
    print(row)
```

```python
print('뉴스 검색어 순위')
for i, j, k in naver_keyword_list[:10]:
    print(i,'위',j,'\n',k)
```

```python
print('연예 스포츠 검색어 순위')
for i, j, k in naver_keyword_list[10:]:
    print(i,'위',j,'\n',k)
```
