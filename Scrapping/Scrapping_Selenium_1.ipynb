## 네이버 영화사이트의 랭킹

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'

driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(3)
driver.get(url)

target_selector = '#old_content > table > tbody > tr > td.title > div > a'
result = driver.find_elements_by_css_selector(target_selector)
print(len(result))

for i in range(10):
    print(f'{i+1}위 {result[i].text}')
```

```python
# 네이버 영화사이트 접속
# 평점순(현재상영영화) 클릭 : click() 메서드 적용
# 순위 정보 추출
# 2차원 리스트로 저장
# 데이터프레임으로 저장
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'

driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(3)
driver.get(url)

target_selector = '#old_content > div.tab_type_6 > ul > li:nth-child(2) > a '
driver.find_element_by_css_selector(target_selector).click()

target_selector1 = '#old_content > table > tbody > tr > td.title > div > a'
target_selector2 = '#old_content > table > tbody > tr > td.point'
result1 = driver.find_elements_by_css_selector(target_selector1)
result2 = driver.find_elements_by_css_selector(target_selector2)

for i in range(10):
    print(f'{i+1}위 {result1[i].text} 평점 {result2[i].text}')
```

```python
naver_movie_ranking = []
for i in range(50):
    naver_movie_ranking.append([result1[i].text, result2[i].text])
    
naver_movie_ranking
```
