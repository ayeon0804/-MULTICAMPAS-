```python
td_list = soup.find_all('td',{'class':'title'})
print(len(td_list))
print(td_list[0])
```

```python
movie_review_list = []
for td in td_list:
    title = td.a.string
    score= td.div.em.string
    review = td.br.next_sibling.strip()
    movie_review_list.append([title, score, review])
```

### 페이지 범위 지정해서 저장하기

```python
page_total = 20
movie_review_page20 = []

for i in range(page_total):
    page_num= i+1
    
    base_url = 'https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=167697&target=after&page='
    url = base_url + str(page_num)
    html = urlopen(url)
    html_str = html.read()
    
    soup = bs(html_str.decode('cp949'), 'html.parser')
    td_list = soup.find_all('td',{'class':'title'})
    
    for td in td_list:
        title = td.a.string
        score= td.div.em.string
        review = td.br.next_sibling.strip()
        movie_review_page20.append([title, score, review])
```

# 다음 페이지로 이동하는 하이퍼링크 선택자 없다면? 마지막페이지.

```python
kid_movie_review = []
page_num = 0 

while True:
    page_num += 1
    
    base_url = 'https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=13558&target=after&page='
    url = base_url + str(page_num)
    html = urlopen(url)
    html_str = html.read()
    
    # soup 객체화 
    # 글자가 깨진다면 decode('cp949/utf-8') 함수 적용 
    soup = bs(html_str.decode('cp949'), 'html.parser')
    td_list = soup.find_all('td', {'class':'title'})
    
    for td in td_list:
        title = td.a.string
        score = td.div.em.string
        review = td.br.next_sibling.strip()
        kid_movie_review.append([title, score, review])
        
    # 마지막 페이지면 종료 
    page_next = soup.find_all('a', {'class':'pg_next'})
    if len(page_next) == 0:
        break
```
