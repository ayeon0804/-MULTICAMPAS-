#  ALEXA 사이트의 Top Sites 페이지

- url : https://www.alexa.com/topsites

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.alexa.com/topsites'
html = urlopen(url)
html_str = html.read()

soup = bs(html_str, 'html.parser')
print(soup)
```

# 참고 html 소스 

<div class="tr site-listing">
                  <div class="td">1</div>
                  <div class="td DescriptionCell">
                    <p class="">
                        <a href="/siteinfo/google.com">Google.com</a>                
                    </p>
                  </div>
                   <div class="td right"><p>15:47</p></div>
                   <div class="td right"><p>17.10</p></div>
                   <div class="td right"><p>0.30%</p></div>
                   <div class="td right"><p>1,300,441</p></div>
</div>


```python
# soup객체.find_all(태그명, {속성:값}) => 리스트

div_list = soup.find_all('div', {'class':'tr site-listing'})
print(len(div_list))
print(div_list[0])
print('\n')
print(div_list[0].find('p'))
print('\n')
print(div_list[0].find_all('a'))
```

```python
print(div_list[0].find('div').string)
print(div_list[0].find('a').string)
print(div_list[0].find_all('p')[1].string)
print(div_list[0].find_all('p')[2].string)
print(div_list[0].find_all('p')[3].string)
print(div_list[0].find_all('p')[4].string)
```

```python
# 리스트로 저장하기 
topsites_list = []
for div in div_list:
    rank = div.find('div').string
    site_name = div.find('a').string
    daily_Time = div.find_all('p')[1].string
    daily_Pageviews = div.find_all('p')[2].string
    daily_Traffic = div.find_all('p')[3].string
    total_Sites_Linking_In = div.find_all('p')[4].string
    topsites_list.append([rank, site_name, daily_Time, daily_Pageviews, daily_Traffic,  total_Sites_Linking_In])
print(len(topsites_list))   
print(topsites_list[0])   
print(topsites_list[-1])
```

```python
# 데이타프레임
field_list = ['Rank', 'Site','Daily Time on Site', 'Daily Pageviews per Visitor', 
              '% of Traffic From Search', 'Total Sites Linking In']
df = pd.DataFrame(topsites_list, columns = field_list)
df
```

```python
# csv 파일로 저장 
df.to_csv('output/topsites.csv', index=False)
```
