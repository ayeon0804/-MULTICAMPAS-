## Color Line

```python
x = np.linspace(0,10,1000)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x))
```

```python
# one of the characters : {'b','g','r','c','m','y','k','w'}
# a string representation of a float value in [0,1] (e.g. '0.5')
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.plot(x, np.sin(x-1), color = 'g')
plt.plot(x, np.sin(x-2), color = '0.5') # gray톤만 가능
plt.plot(x, np.sin(x-3), color = '#FF00FF') # RGB
plt.plot(x, np.sin(x-4), color = (1.0, 0.1, 0.3)) # RGB
plt.plot(x, np.sin(x-5), color = 'tomato')
```

```python
plt.scatter(x, np.tan(x), c = x, cmap = plt.cm.turbo)
```

```python
plt.plot(x, x+0, linestyle = 'solid') # '-'
plt.plot(x, x+1, linestyle = 'dotted') # ':'
plt.plot(x, x+2, linestyle = 'dashed') # '--'
plt.plot(x, x+3, linestyle = 'dashdot') # '-.'
plt.plot(x, x+4, linestyle = '-')
```

```python
plt.plot(x, x+0, color = 'r', linestyle = '-')
plt.plot(x, x+1, '--b')
plt.plot(x, x+2, 'g:')
plt.plot(x, x+3, 'c-.')
```

## Label

```python
plt.plot(x, np.sin(x), label = 'sin(x)')
plt.plot(x, np.cos(x), label = 'cos(x)')
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Pie Chart

```python
labels = ['SciFi','Drama','Thriller','Comedy','Action','Romance']
sizes = [5, 15, 10, 20, 40, 10]
explode = (0, 0, 0, 0, 0.08, 0)
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, explode=explode, startangle=125, shadow=True, autopct='%1.1f%%');
```

```python
Depts = ['COGS','IT','Payroll','R&D','Sales & Marketing']
rp = [30, 15, 25, 10, 20, 30]
ra = [32, 20, 23, 11, 14, 32]
```

```python
plt.subplot(polar=True)
theta = np.linspace(0, 2*np.pi, len(rp))
lines, labels = plt.thetagrids(range(0, 360, int(360/len(Depts))), Depts)
plt.plot(theta, rp)
plt.fill(theta, rp, 'b', alpha=0.1)
plt.plot(theta, ra)
plt.legend(labels=('Plan','Actual'), loc=4);
```
