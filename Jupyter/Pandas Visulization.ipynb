```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts.head(2)
```

```python
ts = ts.cumsum() # 누적
ts.plot();
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index=ts.index,
                 columns=list('ABCD'))
df = df.cumsum()
df.head(2)
```

```python
plt.figure();
df.plot();
```
